```python
from transformers import pipeline
```

```python
##too predict sentence pos or neg

classifier = pipeline("sentiment-analysis")

results = classifier(["i love transformers","this is terrible"])
```

```python
results
```

```python
for i in results:
    print(i['label'])
    print(i['score'])
```

```python
#NER

ner = pipeline("ner",grouped_entities=True)
text = "Elon musk founded spaceX in 2024 and later bough twitter in 2025"
results = ner(text)
results
```

```python
##fill in the blank
##predict the missing word in a sequence

filled_mask = pipeline('fill-mask',model="bert-base-uncased")
sent = "transformers are a very ([MASK]) technology"

predictions = filled_mask(sent)
predictions
```

```python
##zero shot classification
##classify text into user defined categories without training
```

```python
clf =pipeline("zero-shot-classification")
text = "the stock market crashed due to inflation fear and war"
labels = ["finance","sports","health","technology"]

res = clf(text,labels)
res
```

```python
##translations

##from english to french

translator = pipeline("translation_en_to_fr")
output = translator("the weather is nice today")
print(output[0]['translation_text'])
```

```python
##question and answering

qa = pipeline("question-answering")
context = "Hugging face is a company,that provides lot of open sources tools and library models"

question = "what does hugging face do"
result = qa(question,context)
result
```
