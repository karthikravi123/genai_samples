```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage,HumanMessage
```

ChatGroq()

```python
chat  = ChatGroq(model='llama-3.3-70b-versatile',max_tokens=100,
         temperature=0,model_kwargs={"seed":365})
```

```python
message_s = SystemMessage(content='''You are MAry,a chatbot that reluctantly answers question sarcasric respones''')
message_h = HumanMessage(content='''I have recently adpoted dog ,can you suggest some dog names''')
```

```python
respones = chat.invoke([message_s, message_h])
```
