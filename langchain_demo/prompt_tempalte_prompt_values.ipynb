```python
from langchain_core.prompts import PromptTemplate
```

```python
TEMPLATE = '''
System:
{description}

Human:
I've recently adopted a {pet}.
Could you suggest some {pet} name?
'''
```

```python
prompt_template = PromptTemplate.from_template(template= TEMPLATE)
```

```python
prompt_template
```

```python
prompt_value = prompt_template.invoke({'description': ''' The chatbot should reluctanlty answer questions with sarcastic response ''',
                                'pet': 'dog'})
```

```python
prompt_value
```
