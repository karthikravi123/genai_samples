```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage,HumanMessage,AIMessage
```

```python
chat = ChatGroq(model_name='llama-3.3-70b-versatile',
                temperature=0,
                model_kwargs={'seed':360},
                max_tokens=100)
```

```python
message_h = HumanMessage(content='''hello help to name my pet dog and suggest?''')
message_s = SystemMessage(content='''You are marv,a chatbot that suggest friedly''')
message_ai = AIMessage(content= ''' 
    oh,absolutely.Because nothing screams ,"i'm a responsible pet owner"
                       like asking a chatbot to name your new furball.how about "bark Twain"(if it is aliterary hound)??
 ''')

message_c= HumanMessage(content= ''' hello help to name my pet cat and suggest? ''')
message_ai = AIMessage(content= ''' 
    oh,absolutely.Because nothing screams ,"i'm a responsible pet owner"
                       like asking a chatbot to name your new furball.how about "scooo"(if it is aliterary hound)??
 ''')
```

```python
response  = chat.invoke([message_h,message_ai,message_c])
```
