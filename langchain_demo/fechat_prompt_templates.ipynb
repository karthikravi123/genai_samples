##few chat prompt template

```python
import os 
from dotenv import load_dotenv
load_dotenv()
import sys
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import SystemMessagePromptTemplate,HumanMessagePromptTemplate,ChatPromptTemplate,FewShotChatMessagePromptTemplate
```

```python
chat = ChatGroq(
    model_name = 'llama-3.3-70b-versatile',
    temperature=0,
    model_kwargs= {'seed':365},
    max_tokens=100
)
```

```python
TEMPLATE_H = ''' I've recently adopted a {pet}
                could you suggest some {pet} name
    '''

TEMPLATE_AI = ''' {response}'''
```

```python

message_templaet_Ai = SystemMessagePromptTemplate.from_template(template= TEMPLATE_AI )

message_templaet_h = HumanMessagePromptTemplate.from_template(template= TEMPLATE_H )
```

```python
example_template = ChatPromptTemplate.from_messages([message_templaet_Ai,message_templaet_h])
```

```python
examples = [ {'pet':'dog',
             'response': ''' Oh ,abosultely .Because nothing screams,"I'am responsible pet onwer" 
              like asking a chatbot to name your new furball.How about Bark twain (if it's a literay hound )'''},
              {'pet':'cat',
              
                  'response': ''' Oh ,abosultely .Because nothing screams,"I'am responsible pet owner" 
              like asking a chatbot to name your new furball.How about Bark  (if it's a literay hound )'''
                  
              }
              ]
```

```python
few_shot_prompt = FewShotChatMessagePromptTemplate( examples= examples, 
                                                   example_prompt= example_template)
```

```python
chat_template = ChatPromptTemplate.from_messages([few_shot_prompt,message_templaet_h])
```

```python
chat_value = chat_template.invoke({'pet':'''rabbit'''})
```

```python
chat_value.messages
```

```python
for i in chat_value.messages:
    print(type(i))
```

```python
for i in chat_value.messages:
    print(f'{i.type} : {i.content}')
```

```python
response = chat.invoke(chat_value)
```
