```python
import os 
from dotenv import load_dotenv
load_dotenv()
import sys
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import SystemMessagePromptTemplate,HumanMessagePromptTemplate,ChatPromptTemplate
```

```python
chat = ChatGroq(
    model_name = 'llama-3.3-70b-versatile',
    temperature=0,
    model_kwargs= {'seed':365},
    max_tokens=100
)
```

```python
TEMPLATE_S = '{description}'
TEMPLATE_H = ''' I've recently adopted a {pet}
                could you suggest some {pet} name
    '''

message_templaet_s = SystemMessagePromptTemplate.from_template(template= TEMPLATE_S )

message_templaet_h = HumanMessagePromptTemplate.from_template(template= TEMPLATE_H )
```

```python
message_templaet_h
```

##combine both into single prompt template

```python
chat_template = ChatPromptTemplate.from_messages([message_templaet_s,message_templaet_h])
```

```python
chat_template
```

```python
chat_value = chat_template.invoke({'description': ''' The chatbot should reluctantly answer questions
                                   with sarcastic response ''',
                                   'pet': '''dog'''
                                   })
```

```python
chat_value
```

```python
chat_value.messages[0].content
```

```python
response = chat.invoke(chat_value)
print(response.content)
```
