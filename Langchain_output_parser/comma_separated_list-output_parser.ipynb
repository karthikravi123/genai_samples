```python
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import CommaSeparatedListOutputParser 
```

```python
chat = ChatGroq(model_name = 'llama-3.3-70b-versatile')
message_h = HumanMessage(content=''' I've recently adopted a dog.Could you suggest some dog names?
                         { CommaSeparatedListOutputParser().get_format_instructions() }
                          ''')
 
response = chat.invoke([message_h])
```

```python
print(response.content)
```

```python
list_output_parser = CommaSeparatedListOutputParser()
```

```python
respone_parsed = list_output_parser.invoke(response)
```

```python
print(response.content)
```
