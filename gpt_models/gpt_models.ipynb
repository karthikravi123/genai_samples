```python
from openai import OpenAI

client = OpenAI()

def generate_text(prompt):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",   
        messages= [
            {"role":"user","content":prompt}
        ],
        max_tokens=10,
        temperature=0.7
    )
    return response.choices[0].message.content.strip()
```

```python
prompt = "once upon a time"
```

```python
generate_text(prompt)
```

Summarsing text

```python
def text_summarizer(prompt):
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {
                "role":"system",
                "content": "You will be provided block of text and your task is to extract a list of keywords from it"
            },
            {
                "role": "user",
                "content" : "A flying saucer seen by a guest house,a 7ft aline-like figure out of a hedge and a \"cigar-shaped\"  UFO near a school yard.\n"
            },
            {
                "role":"assistant",
                "content": "flying saucer,guest house,7ft alien-like figure,cigars-haped UFO,scholl yard,Uk"
            },
            {
                "role":"user",
                "content": prompt
            },


        ],
        temperature=0.5,
        max_tokens=100,
        )
    return response.choices[0].message.content.strip()
```

```python
prompt = "my name is karthik,im from bangalore completed mtech"
```

```python
text_summarizer(prompt)
```
