```python
from langchain_core.runnables import RunnableLambda
```

```python
runnable_sum = RunnableLambda(lambda x:sum(x))
```

```python
runnable_sum.invoke([2,4])
```

```python
runnbale_Square = RunnableLambda(lambda x: x**2)
```

```python
chain = runnable_sum | runnbale_Square
```

```python
chain.invoke([1,2,3])
```
