```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
RunnablePassthrough().invoke('hi')
```

```python
RunnablePassthrough().invoke([1,2,3])
```

```python
chat_template_tools = ChatPromptTemplate.from_template( ''' 
                        what are the five moist important tolls a {job title} needs?
                        Answer  only by listing the tools                               
                                                        ''')

chat_template_startegy = ChatPromptTemplate.from_template( 
    ''' COnsidering the tools provided,develop a strategy for effectively learning and mastering them
    {tools}
    '''
)
```

```python
chat_template_tools
```

```python
chat = ChatGroq(model_name='llama-3.3-70b-versatile',
                temperature=0,
                model_kwargs={'seed':360},
                #api_key=os.environ["GROQ_API_KEY"],

                max_tokens=100)
```

```python
string_parser = StrOutputParser()
```

```python
chain_tools = chat_template_tools | chat | string_parser

chain_Strategy = chat_template_startegy| chat |string_parser
```

```python
print(chain_tools.invoke({'job title': 'data scientist'}))
```

```python
print(chain_Strategy.invoke({'tools':'''1. Python
2. R
3. SQL
4. Excel
5. Tableau '''}))
```

```python
chain_tools = chat_template_tools | chat | string_parser | {'tools' : RunnablePassthrough()}
chain_Strategy = chat_template_startegy | chat | string_parser
```

```python
print(chain_tools.invoke({'job title': 'data scientist'}))
```

```python
chain_combined = chain_tools | chain_Strategy
```

```python
chain_combined.invoke({'job title':'data scientist'})
```
