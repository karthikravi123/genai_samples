```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import CommaSeparatedListOutputParser
```

```python
chat_template = ChatPromptTemplate.from_messages([('human',"I've adopted a {pet} which is a {breed}.could you suggest several raining tips.?")])
```

```python
chat = ChatGroq(model_name='llama-3.3-70b-versatile',
                temperature=0,
                model_kwargs={'seed':360},
                #api_key=os.environ["GROQ_API_KEY"],

                max_tokens=100)
```

```python
chain = chat_template | chat
```

```python
chain.invoke({'pet':'dog','breed':'shepered'})
```

```python
chain.batch([{'pet':'dog','breed':'shpered'},{'pet':'dragon','breed': 'night furry'}])
```
