```python
from langchain_core.runnables import RunnableLambda
from langchain_core.runnables import chain
```

```python
def find_sum(x):
    return sum(x)

def find_square(x):
    return x**2
```

```python
chain = RunnableLambda(find_sum) | RunnableLambda(find_square)
```

```python
chain.invoke([1,2,3])
```

```python
@chain
def find_sum(x):
    return sum(x)

@chain
def find_square(x):
    return x**2
```

```python
type(find_sum)
```

```python
chain = find_sum | find_square
```

```python
chain.invoke([1,2,3])
```
