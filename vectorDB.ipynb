```python
import faiss
#facebook ai similarity search
```

```python
# !pip install sentence-transformers
```

```python

import numpy as np
from sentence_transformers import SentenceTransformer  ##turning text intoembedding
##transormer based from huggung face
#model used for embedding
```

```python
model = SentenceTransformer("all-MIniLM-L6-v2")
print(model)
```

```python
documents = ['The cat is ont he mat',
             'Dog is mans good friend',
             'The quick brown fox jumps',
             'i love playing football every weekend'
             ]

print(documents)
```

```python
embeddings = model.encode(documents,show_progress_bar=True)
embeddings
```

```python
embeddings[0]
```

```python
#faiss index creation
#convert the embedding tp float32

embeddings_np = np.array(embeddings).astype("float32")
embeddings_np
```

```python
embeddings_np.shape
```

```python
embeddings_np.shape[1]
```

```python
#create index
dimension = embeddings_np.shape[1]
dimension
```

```python
index = faiss.IndexFlatL2(dimension)
```

```python
##add embeddings to index
index.add(embeddings_np)
print("Faiss index size -",index.ntotal)
```

```python
###searching similar items
query = "I enjoy playing soccer"

query_vector = model.encode([query]).astype('float32')
print(query_vector)
```

```python
print(len(query_vector))
```

```python
##number of nearest neighbours (k)

k = 3
distances,indices = index.search(query_vector,k)
print("top 3 results")
for i in indices[0]:
  print(documents[i])
```

```python
print(indices)
print(distances)
```

```python
##save the index
faiss.write_index(index,'faiss_index.index')
```

```python
#load the index
loaded_index = faiss.read_index('faiss_index.index')
loaded_index
```

```python
import chromadb
```

```python
client = chromadb.Client()
```

```python
collection = client.get_or_create_collection(name="my_documents")
collection
```

```python
documents = ['The cat is ont he mat',
             'Dog is mans good friend',
             'The quick brown fox jumps',
             'i love playing football every weekend'
             ]
```

```python
embeddings = model.encode(documents,show_progress_bar=True)
embeddings
```

```python
###adding the documents in my collections
collection.add(documents=documents,
               ids = [str(i) for i in range(len(documents))],embeddings=embeddings.tolist())
```

```python
query = "i like to play soccer"
query_embedding = model.encode([query]).tolist()

results = collection.query(query_embeddings=query_embedding,n_results=2)
results
```
